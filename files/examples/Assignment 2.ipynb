# Assignment 2: Learn interrupts with button

This notebook demonstrates how to read button input and create interactive LED responses.

```micropython
import machine
import time
from neopixel import NeoPixel
```

```micropython
# Configuration of NeoPixel LED
led_pin = machine.Pin(2, machine.Pin.OUT)
np = NeoPixel(led_pin, 1)  # 1 LED

def set_color(r, g, b):
    """Set LED to specified RGB color"""
    np[0] = (r, g, b)
    np.write()

def clear_led():
    """Turn off the LED"""
    np[0] = (0, 0, 0)
    np.write()

# Test the LED
set_color(0, 255, 0)  # Green
time.sleep(0.5)
clear_led()
print("LED test complete!")
```

```micropython
# Configure button on pin 9 with pull-up resistor
button = machine.Pin(9, machine.Pin.IN, machine.Pin.PULL_UP)
print("Button configured on pin 9")
```

```micropython
def wait_for_button_press():
    """Wait until button is pressed"""
    while button.value() == 1:  # Wait while button is not pressed (HIGH with pull-up)
        time.sleep(0.01)
    time.sleep(0.05)  # Debounce delay

# Test it - press the button!
print("Waiting for button press...")
wait_for_button_press()
print("Button pressed!")
```

```micropython
led_state = False

print("Press button to toggle LED. Press Ctrl+C to stop.")

try:
    while True:
        wait_for_button_press()
        led_state = not led_state
        
        if led_state:
            set_color(0, 255, 0)  # Green when ON
            print("LED ON")
        else:
            clear_led()
            print("LED OFF")
        
        # Wait for button release
        while button.value() == 0:
            time.sleep(0.01)
        time.sleep(0.05)  # Debounce
        
except KeyboardInterrupt:
    clear_led()
    print("\nStopped by user")
```

```micropython
def measure_press_duration():
    """Measure how long the button is held down"""
    start_time = time.time()
    while button.value() == 0:  # While button is pressed (LOW)
        time.sleep(0.01)
    end_time = time.time()
    return end_time - start_time

print("Press and hold the button...")
wait_for_button_press()
duration = measure_press_duration()
print(f"Button held for {duration:.2f} seconds")
```

## Bonus Challenge: Visual Feedback Based on Duration

Change LED color based on how long the button is pressed:
- Short press (< 1s): Blue
- Medium press (1-3s): Yellow
- Long press (> 3s): Red

**Note:** Press Ctrl+C to stop.

```micropython
print("Press button for different durations. Press Ctrl+C to stop.")

try:
    while True:
        wait_for_button_press()
        duration = measure_press_duration()
        
        if duration < 1.0:
            set_color(0, 0, 255)  # Blue - short press
            print(f"Short press: {duration:.2f}s - Blue")
        elif duration < 3.0:
            set_color(255, 255, 0)  # Yellow - medium press
            print(f"Medium press: {duration:.2f}s - Yellow")
        else:
            set_color(255, 0, 0)  # Red - long press
            print(f"Long press: {duration:.2f}s - Red")
        
        time.sleep(1)  # Show color for 1 second
        clear_led()
        time.sleep(0.2)  # Brief pause before next press
        
except KeyboardInterrupt:
    clear_led()
    print("\nStopped by user")
```
